```python
import re
def clean_txt(text):
    text = re.sub(r'[\t"\']', '', text) 
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) 
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
```

```python
data['description'] = data['description'].apply(clean_txt)
```

Remove stopwords..........>>>>>>

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = text.split()
    return ' '.join([word for word in words if word not in stop_words])
```

```python
data['description'] = data['description'].apply(remove_stopwords)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['description'] = data['description'].apply(stem)
```

```python
model = load_model("disease_prediction_model.h5")

df = pd.read_csv("doctorg_embedding.csv")
label_encoder = LabelEncoder()
df["disease_encoded"] = label_encoder.fit_transform(df["name"])

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
model_llm = genai.GenerativeModel("models/gemini-2.0-flash-exp")

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")


def extract_symptoms(natural_language_input):
    """
    Extracts a list of symptoms mentioned in the input and removed stopwords using spacy.
    """
    prompt = f"""Please extract the symptoms mentioned in the following text and return them as a comma-separated list:

    \"{natural_language_input}\"

    Symptoms:"""
    response = model_llm.generate_content(prompt)
    symptom_string = response.text.strip()
    symptoms = [s.strip() for s in symptom_string.split(',')]
    processed_symptoms = []
    for symptom in symptoms:
      doc = nlp(symptom.lower())
      filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
      processed_symptoms.append(" ".join(filtered_tokens).strip())
    return [s for s in processed_symptoms if s]

def predict_disease(user_symptoms):
    sentence_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
    user_embedding = sentence_model.encode(" ".join(user_symptoms)).reshape(1, -1)

    predictions = model.predict(user_embedding)[0]
    top_5_indices = np.argsort(predictions)[-5:][::-1]
    top_5_diseases = [label_encoder.classes_[i] for i in top_5_indices]
    top_5_probs = [predictions[i] for i in top_5_indices]

    return list(zip(top_5_diseases, top_5_probs))

def get_disease_description(disease_name, dataframe):
    """
    Retrieves the description of a given disease from the DataFrame,
    processes it using Google GenAI, and prints the enhanced result.
    """
    disease_row = dataframe[dataframe['name'] == disease_name]
    
    if not disease_row.empty:
        description = disease_row['description'].iloc[0]
    else:
        description = f"Description for '{disease_name}' not found."
        return description

    # Process the description using GenAI
    prompt = f"""Enhance the following disease description to be more informative, clear, and user-friendly:
    
    Disease: {disease_name}
    Original Description: {description}
    
    Enhanced Description:"""
    
    response = model_llm.generate_content(prompt)
    enhanced_description = response.text.strip()
    
    print("\nEnhanced Disease Description:")
    print(enhanced_description)
    
    return enhanced_description

# User input (replace with actual user input if needed)
# user_input ="I am feeling depressed and anxious and bit nervousness and sometimes insomania and palpitations and little bit of dizzy and breathing fast, irregular heartbeat, depressive psycotic symptoms and shortness of breath and dizziness"
user_input = " I am having hoarse voice, sore throat, difficulty in speaking, cough, throat swelling, skin swelling, and retention in urine."
# user_input = "I am having fatty liver, Constipation,  death"
symptom_arr = extract_symptoms(user_input)
print("Extracted Symptoms:", symptom_arr)

predicted_diseases = predict_disease(symptom_arr)

print("\nBased on these symptoms, the top 5 most likely diseases are:")
if predicted_diseases:
    for i, (disease, probability) in enumerate(predicted_diseases):
        confidence_percentage = probability * 100
        print(f"- {disease} with a probability of {confidence_percentage:.2f}%")

    top_disease = predicted_diseases[0][0]
    description = get_disease_description(top_disease, df)
    print(description)
  
else:
    print("Could not predict any disease based on the provided symptoms.")


     
```
